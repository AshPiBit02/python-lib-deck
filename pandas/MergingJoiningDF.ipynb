```python
import pandas as pd

df1 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "name": ["Alice", "Bob", "Charlie", "David"],
    "age": [25, 30, 35, 40]
})
df2 = pd.DataFrame({
    "id": [3, 4, 5, 6],
    "department": ["HR", "IT", "Finance", "Marketing"],
    "salary": [50000, 60000, 55000, 45000]
})
```

# Merge

```python
# Merge df1 & df2 on id (default inner join)
pd.merge(df1,df2,on="id")
```

```python
# Perform a left merge of df1 with df2 on id.
pd.merge(df1,df2,on="id",how="left")
```

```python
# Perform a right merge of df1 with df2 on id.
pd.merge(df1,df2,on="id",how="right_on=")
```

# Concat

```python
# Concatenate df1 & df2 vertically(stack rows).
pd.concat([df1,df2],axis=0)
```

```python
# Concatenate df1 & df2 horizontally.
pd.concat([df1,df2],axis=1)
```

```python
# Concatenate df1 twice and reset the index.
pd.concat([df1,df1],ignore_index=True)
```

# Join

```python
# Set id as index in both DF and join them.
df1=df1.set_index("id")
df2=df2.set_index("id")
df1.join(df2,how="outer")
```

```python
# Perform a left join.
df1.join(df2,how="left")
```

```python
# Perform an inner join between df1 & df2 on id.
df1.join(df2,on="id",how="inner")
```
