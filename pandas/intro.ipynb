Pandas is a Python library built on top of NumPy, providing high-level data stuctures and opeations for manipulating numerical tables and time series.
 
 Core Structures:
 
   Series -> One-dimensional labeled array.
   DataFrame -> Two-dimensional labeled table(rows and columns).

 Why Pandas are used:

   Data Cleaning -> Handle missing values, duplicates, and inconsistent data.
   Data Transformation -> Merge, filter, reshape, and aggregate datasets easily.
   Exploratory Data Analysis(EDA) -> Generate descriptive statistics, slice/filter data, and explore trends.
   Integration -> Works seamlessly with NumPy, Matplotlib, Seaborn, SciPy, and Scikit-learn.
   Time Series Handling -> Specialized tools for resampling, rolling averages, and working with time-indexed data.

 Use Cases:

   Data Wrangling & Cleaning
     -> Preparing raw datasets for Analysis
   Exploratory Data Analysis
     -> Quick insights into data distributions and relationships.
   Data Transformation
     -> Pivot tables, group-by operations, reshaping datasets.
   Time Series Analysis
     -> Financial data, sensor readings, event logs.
   Data Export/Integration
     -> Save results to CSV, Excel, SQL, or JSON for sharing.

  Applications:

   -> Data Science & Machine learining(Preprocessing datasets before modeling.)
   -> Finance(Stock market Analysis, risk modeling, portfolio tracking.)
   -> Business Analysis(Customer segmentation, sales forecasting, KPI dashboards.)
   -> Healthcare(Patient record Analysis, medicine research datasets.)
   -> Web/Data Engineering(Log Analysis, ETL pipelines.)
   -> Academia & Research(Handling large datasets in scientific studies.)

Series

  -> A one-dimensional array-like object with labels(index).
  
  Data Types(Homogeneous(usually one type)): Can hold integers, floats, strings, or mixed types.


```python
import pandas as pnd 
s=pnd.Series([10,20,13],index=["a","b","c"])
print(s)
print(s["a"])
```

DataFrame

 -> A two-dimensional, size-mutable, heterogeneours tabular data structure with rows and columns.

Composition:

  -> Each column in a DataFrame is a Series.

Data Types(Heterogeneous)

  -> can store differenct types per column

```python
data={
    "Name":["Aashish","Jack","Rome"],
    "Age":[21,22,19],
    "Marks":[93,85,99]
}
df=pnd.DataFrame(data)
print(df)
```
