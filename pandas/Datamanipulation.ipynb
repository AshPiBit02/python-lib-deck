```python
import pandas as pd

data = {
    "ProductID": [101, 102, 103, 104, 105],
    "Product": ["Laptop", "Phone", "Tablet", "Monitor", "Keyboard"],
    "Category": ["Electronics", "Electronics", "Electronics", "Electronics", "Accessories"],
    "Price": [800, 600, 300, 200, 50],
    "Stock": [10, 25, 15, 8, 50]
}

df = pd.DataFrame(data)
df=df.set_index("ProductID")
print(df)
```

# Add New Column

```python
# Add a column Discount=10% of Price.
df["Discount"]=df["Price"]*0.1
print(df)
```

```python
# Add a column StockValue=Price * Stock
df["StockValue"]=df["Price"]*df["Stock"]
print(df)
```

# Delete Column

```python
# Remove the Category Column
df=df.drop(columns=["Category"])
print(df)
```

```python
# Remove both ProductID and Stock.
df=df.drop(columns=["Product","Stock"])
print(df)
```

# Apply Function

```python
# Apply a Function to Price taht adds 13% VAT.
df["Price"]=df["Price"].apply(lambda x : x*1.13)
print(df)
```

```python
# Apply a function to Stock that labels "Low" if < 10, else "High".
df["Stock"]=[10,25,15,8,50] # for operation
df["Label"]=df["Stock"].apply(lambda x: "Low" if x<10 else "High")
print(df)

```

# Map Function

```python
df=pd.DataFrame(data)
# Map Product names to their lengths 
df["Product"]=df["Product"].map(len)
print(df)
```

```python
# Map Category values: "Electronice" -> "Elec", "Accessories" -> "Acc".
df["Category"]=df["Category"].map({"Electronics":"Elec","Accessories":"Acc"})
print(df)
```

# Sorting Values

```python
# Sort products by Price(descending).
# print(df.sort_values("Price",ascending=False))
print(df.sort_values(by="Price",ascending=False))
# print(df.sort_values(by=["Price","Stock"],ascending=[False,True])) # for multiple columns sorting
```

```python
# Sort products by Stock(ascending)
print(df.sort_values("Stock"))
# print(df.sort_values(by="Stock",ascending=True))
```

# Sorting Index

```python
# Reverse the DataFrame index order
df=df.sort_index(ascending=False)
print(df)
```

```python
# Sort by index after shuffling rows
df=df.sample(frac=1) # sample(frac=1) reorders rows randomly
print(df)

df=df.sort_index()
print(df)
```
