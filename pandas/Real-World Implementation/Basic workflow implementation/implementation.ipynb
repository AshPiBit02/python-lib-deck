```python
import pandas as pd
df=pd.read_csv("sales_data.csv",)
print(df)
```

```python
# Type of data present
print(df.info(memory_usage="deep"))
```

```python
# downcasting for optimization
df[df.select_dtypes(include="int64").columns]=df.select_dtypes(include="int64").apply(pd.to_numeric,downcast="integer")
print(df.info(memory_usage="deep"))
```

```python
# check if values is null then drop the record containing that null value
print(df.isnull())
```

# Business Insights

```python
total_revenue=(df["price"]*df["quantity"]).sum()
print(f"Total Revenue: {total_revenue}$")
```

```python
# most sold product
product_sales=df.groupby("product")["quantity"].sum()
max_qty=product_sales.max()
most_sold_product=product_sales[product_sales==max_qty]
print(most_sold_product)
```

```python
# most profitable category
df["revenue"]=(df["price"]*df["quantity"]).sum()
category_sales=df.groupby("category")["revenue"].sum()
max_revenue=category_sales.max()
most_profitable_category=category_sales[category_sales==max_revenue]
print(most_profitable_category)
```

```python
# City with high sales
city_sales=df.groupby("city")["revenue"].sum()
# print(f"City: {city_sales.idxmax()}\nSales: {city_sales.max()}")

max_sales=city_sales.max()
city_with_hight_sales=city_sales[city_sales==max_sales] # handles multiple max sales cities
print(city_with_hight_sales)
```

```python
# most perferred payment method
payment_methods=df.groupby("payment_method").size()
print(payment_methods.idxmax())
```

# Customer Analysis

```python
# Average age of customer according to city
average_cutomer_age=df.groupby("city")["age"].mean()
print(average_cutomer_age)
```
