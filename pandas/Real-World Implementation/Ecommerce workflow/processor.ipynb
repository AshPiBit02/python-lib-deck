```python
# Load Data
import pandas as pd
df=pd.read_csv("ecommerce_data.csv")
```

# Clean Data

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df.drop_duplicates(inplace=True)
df.tail()
```

# Transform Data

```python
# Total revenue per order
df["total_amount"]=df["price"]*df["quantity"]
pos=df.columns.get_loc("quantity")+1 # find index of quantity
df.insert(pos,"total_amount",df.pop("total_amount"))
df=df.set_index("order_id")
print(df.head())
```

```python
# orderdate to datetime
df["order_date"]=pd.to_datetime(df["order_date"])
print(df.dtypes)
```

# Analyze Data

```python
# Total Revenue
total_renvenue=df["total_amount"].sum()
print(f"Total Revenue: {total_renvenue}$")
```

```python
# Most sold product
result=df.groupby("product")["quantity"].sum().sort_values(ascending=False).head(3) # top 3 product
print(f"Most sold Products: \n{result}")
```

# Group Data

```python
# Revenue by Category
revenue_by_category=df.groupby("category")["total_amount"].sum()
print(revenue_by_category)
```

```python
# Sales by City
sales_by_city=df.groupby("city")["total_amount"].sum()
print(sales_by_city)
```
