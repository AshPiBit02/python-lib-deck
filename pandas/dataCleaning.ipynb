```python
import pandas as pd

data = {
    "Id": [1, 2, 3, 4, 5],
    "Name": ["Alice", "Bob", None, "David", "Eva"],
    "Age": [23, None, 45, 29, None],
    "Department": ["HR", "IT", "Finance", "IT", "HR"],
    "Salary": [30000, 55000, None, 48000, 35000],
    "Experience": [1, 7, 15, None, 4]
}

df = pd.DataFrame(data)
df=df.set_index("Id")
print(df)
```

# Handling Missing Values

```python
# Finding which cells are missing
print(df.isnull())
print(df[df["Name"].isnull()]) # shows data whose name is null
```

```python
# Count total missing values in each column.
result=df.isnull().sum()
print(result)
```

# Dropping Missing Data

```python
# Drop all rows where any value is missing.
result=df.dropna()
print(result)
```

```python
# Drop only rows where Salary is missing.
result=df.dropna(subset=["Salary"]) # subset tell pandas to only check the salary column
print(result)
```

# Filling Missing Data

```python
# Fill missing Age values with the average age.
df["Age"]=df["Age"].fillna(df["Age"].mean())
print(df)
```

```python
# Fill missing Experience values with 0.
df["Experience"]=df["Experience"].fillna(0)
print(df)
```

# Replacing Values

```python
# Replace all "HR" in Department with "Human Resources".
df["Department"]=df["Department"].replace("HR","Human Resources")
print(df)
```

```python
# Replace all None in Name with Unknown.
df["Name"]=df["Name"].fillna("Unknown")
print(df)
```

# Removeing Duplicates

```python
# Add a duplicate row manually, then remove duplicates
df.loc[5]=df.loc[1]
print(df)
```

```python
df=df.drop_duplicates()
print(df)
```

# Renaming Columns

```python
# Rename salary -> MonthlySalary
df=df.rename(columns={"Salary":"MonthlySalary"})
print(df)
```

```python
# Rename Department -> Dept
df=df.rename(columns={"Department":"Dept"})
print(df)
```

# Cleaining Data Types

```python
# Convert Age column to int
df["Age"]=df["Age"].astype("int64") # allows int + NaN
print(df.dtypes)
```

```python
# Convert MonthlySalary column to float
df["MonthlySalary"]=df["MonthlySalary"].astype("float16")
print(df.dtypes)
```
