```python
import pandas as pd

data = {
    "Id": range(1, 11),
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", 
             "Frank", "Grace", "Helen", "Ian", "Jane"],
    "Age": [23, 35, 45, 29, 31, 40, 22, 28, 36, 27],
    "Department": ["HR", "IT", "Finance", "IT", "HR", 
                   "Finance", "IT", "HR", "Finance", "IT"],
    "Salary": [30000, 55000, 70000, 48000, 35000, 
               65000, 40000, 37000, 72000, 50000],
    "Experience": [1, 7, 15, 5, 4, 12, 2, 3, 10, 6]
}

df = pd.DataFrame(data)
df=df.set_index("Id") # set Id as index
print(df)
```

```python
print(df["Name"].to_string(index=False)) # to_string() helps to restrict display index
```

```python
print(df.iloc[:3])
```

```python
print(df.loc[2:5])
```

```python
print(df[df["Age"]>30])
```

```python
print(df[(df["Department"]=="IT") & (df["Salary"]>45000)])
```

```python
print(df["Salary"].tail(2))
```

```python
print(df.loc[3:7,["Name","Department"]])
```

```python
print(df[(df["Department"]=="Finance") & (df["Experience"]>=10)])
```

```python
print(df.iloc[3:8]) #exclusive
```
