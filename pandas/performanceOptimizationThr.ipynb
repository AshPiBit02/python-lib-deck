# Performance Optimization
 1. Vectorization 
   * It means applying operations to entire arrays/columns at once, instead of looping row by row.
   * Use case: Adding two columns, computing ratios, or applying math functions across millions of rows instantly.
   * Essential methods/elements:
      * Direct arithmetic: df["a"]+df["b](instead of looping).
      * Universal functions: np.log(df["a"]),np.sqrt(df["b"]).
      * Avoid .apply() with Python functions unless necessary
 2. Avoid Loops
   * Python for loops are slow in DF because they iterate row by row. Pandas methods are optimized to handle entire columns.
   * Use cases: Instead of looping to calculate percentages, use vectorized division.
   * Essential methods/elements:
      * .map() for element-wise transformation on Series.
      * .applymap() for element-wise transformation on DF.
      * .assign() to add new computed columns without loops.
      * .query() for filtering with expressions instead of looping.
 3. .values
   * .values gives you the underlying Numpy array to a DF or Series.
   * Use case: When you need raw numerical arrays for fast computation or to pass data into ML models.
   * Essential methods/elements:
      * df.values -> returns a 2D NumPy array.
      * series.values -> returns a 1D NumPy array.
      * Fast, but loses column/row labels.
 4. .astype()
   * Changes the data type of a column.
   * Use case: converting floats to integers, or strings to categories for memory efficiency.
   * Essential methods/elements: 
      * df["col"].astype("int32) -> smaller integer type.
      * df["col].astype("float32") -> smaller float type.
      * df["col"].astype("category") -> categorical encoding.
 5. Memory Optimization
   * Reducting the memory footprint of large DF by using efficient dypes.
   * Use case: Handling datasets with millions of rows without running out of RAM.
   * Essential methods/elements:
      * pd.to_numeric(df["col"],downcast="integer") -> downcast to smallest integer type.
      * pd.to_numeric(df["col"],downcast="float") -> downcast to smallest float type.
      * df.info(memory_usage="deep") -> check memory usage.
      * Convert object/string columns to category when values repeat.
 6. Category dtype
   * A special dtype for columns with repeated string values. Stores them as integer codes internally.
   * Use case: Columns like "Region" or "Product" with limited unique values but millions of rows.
   * Essential methods/elements:
      * df["col"].astype("category") -> convert to category.
      * df["col"].cat.codes -> view integer codes.
      * Saves memory and speeds up groupby/joins.
