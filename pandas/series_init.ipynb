# Components
1. Data
  * The actual vlaues(integers,floats,strings,objects).
2. Index
  * Labels for each element(defult is 0,1,2,....).
3. dtype
  * Data type of the values
4. name
  * Optional label for the Series.

# Essentail Functions & Methods 
1. Creation
  * pd.Series(data) -> create from list, NumPy array, dict, or scalar.
  * Parameters -> data, index, dtype, name.
2. Accessing Elements
  * By posisition -> s.iloc[0]
  * By label -> s.loc['a']
  * Slicing -> s[1:3],s[['a','b']]
  * Head/Tail -> s.head(3), s.tail(2)
3. Index & Attributes
  * s.index -> Index object.
  * s.values -> Numpy array of values.
  * s.dtype -> Data type.
  * len(s) -> Number of elements.
  * s.name -> Series name.
4. Filtering & Selection
  * Boolean indexing -> s[s>10]
  * .isin() -> Check membership.(Shrinks Series(only matches))
  * .where() -> Condition selection.(Keeps full Series)
5. Statical & Math Operations
  * s.sum(), s.mean(), s.median(), s.std()
  * s.min(), s.max()
  * s.value_counts() -> Frequency of values.
  * s.describe() -> Summary statics.
6. Tranformation 
  * .apply(func) -> Apply custom function
  * .map(func) -> Element-wise mapping.
  * .transform(func) -> Return transformed Series.
  * .cumsum(), .cumprod(), .diff()
7. Handling Missing Data
  * .isnull(), .notnull()
  * .fillna(value) -> Replace missing values.
  * .dropna() -> Remove missing values.
8. Combining Series
  * Arithmetic operations align by index -> s1+s2
  * .add(), .sub(), .mul(), .div() with fill_value for missing indexes.
  * pd.concat([s1,s2]) -> Concatenate Series.
9. Advanced s
  * .sort_values(), .sort_index()
  * .rolling(window).mean() -> Rolling statistics.
  * .shift() -> Shift values by index.

```python
import pandas as pd

# Create a Series with all its compontents
s=pd.Series(
    data=[11,22,33,44,77],
    index=["i","ii","iii",'iv',"v"],
    dtype="int64",
    name="Scores"
)

# Display the Series
print(s)
```

```python
# Access components
print("Values: ",s.values)
print("Index: ",s.index)
print("Data Type: ",s.dtype)
print("Name: ",s.name)
print("Length: ",len(s))
```

```python
# Accessing Elements
s=pd.Series(
    [95,88,76,100,67],
    index=["Math","Physics","Chemistry","English","DSA"],
    name="Exam Scores"
)

print(s.iloc[0]) # used to access data using index position
print(s.loc["Chemistry"]) # used to access data using defined index value

print(s[1:3])
print(s[["Physics","English"]])

print(s.head(2)) # prints first two data
print(s.tail(2)) # prints last two data
```

```python
# Filtering & Selection 
s=pd.Series(
    data=[17,18,11,13,8,37],
    index=["Joh","Rob","Arya","Sansa","Bran","Ned"],
    dtype="int64",
    name="Name_Age"
)
print([s>10])
print(s[s>10])

mask=s.isin([79,37])
print(mask)
print(s[mask])

filter=s.where(s>=15)
print(filter)
```

```python
# Statical & Math Operations
scores = pd.Series(
    [95, 88, 76, 100, 67, 82, 91, 73, 85, 90, 
     78, 84, 69, 99, 72, 81, 65, 87, 93, 70],
    index=[
        "Math", "Physics", "Chemistry", "English", "DSA",
        "Biology", "Statistics", "Economics", "History", "Geography",
        "Computer", "Philosophy", "Art", "Music", "Sports",
        "Psychology", "Sociology", "Law", "Medicine", "Engineering"
    ],
    name="Exam Scores"
)

print("Total marks scored ",scores.sum())

print("Average marks obtained: ",scores.mean())

print("Highest Obtained mark: ",scores.max())

print("Lowest Obtained mark: ",scores.min())

print(scores.value_counts())

print(scores.describe())
```

```python
scores=pd.Series(
    [95,88,76,100,67,82,91,73,85,90],
    index=["Math","Physics","Chemistry","English","DSA","Biology","Statistics","Economics","History","Geography"],
    name="Exam Scores"
)

print(scores.apply(lambda x: x+10))
```

```python
print(scores.map(lambda x:x**2))

```

```python
print(scores.transform(lambda x:x/x.max()))
```

```python
print(scores.cumsum())
```

```python
scores=pd.Series(
    [95,88,76,100,67,82,91,73,85,90],
)
print(scores.diff())
```

```python
print(scores.pct_change())
```

```python
# Combining Series
import pandas as pd
scores1=pd.Series(
    [95,88,76,100,67],
    index=["Math","Physics","Chemistry","English","DSA"],
    name="Midterm Scores"
)
scores2=pd.Series(
    [90,85,80,95,70],
    index=["Math","Physics","Chemistry","English","DSA"],
    name="Final Scores"
)

print(scores1+scores2)
```

```python

print(scores1.sub(scores2))
```

```python
print(scores2.add(scores1))
```

```python
print(scores1.mul(scores2))
```

```python
print(scores1.div(scores2))
```

```python
print((scores1+scores2)/2)
```

```python
scores3=pd.concat([scores1,scores2])
print(scores3)
```

```python
combined_fill=scores1.add(scores2,fill_value=0) # add() helps to prevent NaN
print(combined_fill)
```

```python
print(combined_fill.sort_values())
print(combined_fill.sort_index())
```

```python
import pandas as pd
expenses=pd.Series(
    [1200,800,950,1100,700,1300,900,100,850,1250],
    index=["Rent","Health","Transport","Utilities","Entertainment","Shopping","Health","Education","Travel","Savings"],
    name="Monthly Expenses"
)
print(expenses.info())
print(expenses.rolling(window=4).mean()) # gives the average of every n consecutive categories.

print(expenses.shift(-1)) # values shifts
print(expenses.shift(2))
```
