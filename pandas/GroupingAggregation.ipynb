```python
import pandas as pd

data = {
    "OrderID": [1, 2, 3, 4, 5, 6, 7],
    "Product": ["Laptop", "Facemask", "Pen", "Laptop", "Book", "Phone", "Keyboard"],
    "Category": ["Electronics", "Cosmetics", "Stationary", "Electronics", "Stationary", "Electronics", "Accessories"],
    "Region": ["North", "South", "North", "East", "East", "South", "North"],
    "Quantity": [2, 5, 3, 1, 4, 2, 10],
    "Price": [800, 600, 300, 850, 200, 650, 50]
}

df = pd.DataFrame(data)
df=df.set_index("OrderID")
print(df)
```

# Groupby()

```python
# Group the data by Category and show the total Quantity sold in each category.
print(df.groupby("Category")["Quantity"].sum())
```

```python
# Group the data by Region and show the average Price in each region.
print(df.groupby("Region")["Price"].mean())
```

```python
# Group the data by Product and show the total Quantity sold for each product.
print(df.groupby("Product")["Quantity"].sum())
```

# Aggregation Functions

```python
# Find the sum of Quantity for each Region.
print(df.groupby("Region")["Quantity"].sum())
```

```python
# Find the mean Price for each Category.
print(df.groupby("Category")["Price"].mean())
```

```python
# Find the count of orders for each Product.
print(df.groupby("Product")["Quantity"].sum())
```

# sum(), mean(), count()

```python
# Use .sum() to calculate total Quantity grouped by Region.
print(df.groupby("Region")["Quantity"].sum())
```

```python
# Use .mean() to calculate average Price grouped by Product.
print(df.groupby("Product")["Price"].mean())
```

```python
# Use .count() to count how many orders are in each Region.
print(df.groupby("Region")["Quantity"].count()) 
# print(df.groupby("Region").size()) # size() counts number of rows
```

# Multiple Aggregation

```python
# Group by Region and apply sum, mean, and count on Quantity.
print(df.groupby("Region")["Quantity"].agg(["sum","mean","count"]))
```

```python
# Group by Category and apply sum and mean on Price.
print(df.groupby("Category")["Price"].agg(["sum","mean"]))
```

```python
# Group by Product and apply sum and count on Quantity.
print(df.groupby("Product")["Quantity"].agg(["sum","count"]))
```

# Group by Multiple Columns

```python
# Group by Region and Category to find total Quantity.
print(df.groupby(["Region","Category"])["Quantity"].sum())
```

```python
# Group by Region and Product to find average Price.
print(df.groupby(["Region","Product"])["Price"].mean())
```

```python
# Group by Category and Product to count number of orders.
df=df.reset_index()
print(df.groupby(["Category","Product"])["OrderID"].count())
df=df.set_index("OrderID")
```
