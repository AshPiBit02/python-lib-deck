```python
import pandas as pd
df=pd.read_csv("dataFiles/dummy_sales.csv")
df=df.set_index("order_id")
```

```python
# Vectorized multiplication(Introducing sales amount from item and rate)
df["sales"]=df["quantity"]*df["rate"]
print(df[["item","rate","sales"]].head())
print(df.dtypes)
```

```python
print("Before Downcast:\n",df.info(memory_usage="deep"))
```

```python
# Downcast all numeric columns
df[df.select_dtypes(include=["int64","float64"]).columns]=df.select_dtypes(include=["int64","float64"]).apply(pd.to_numeric,downcast="float")
print("After downcast:\n",df.info(memory_usage="deep"))
```

```python
# Category dtype
df["payment_method"]=df["payment_method"].astype("category")
print(df["payment_method"].cat.categories)
df["payment_method"]=df["payment_method"].cat.codes
print(df[["item","rate","sales","payment_method"]].head())
```

```python
# filter
print(df[["item","sales"]].query("sales>20000").head(11))
```
