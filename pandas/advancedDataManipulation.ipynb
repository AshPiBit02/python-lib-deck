# Advanced Data Manipulation

Functions & Methods
 1. pd.pivot_table() -> creates a pivot table summarizinig data.
     * Great for aggregation across categories.
 2. pd.crosstab() -> Computes a cross-tabulation of two(or more factors).
     * Useful for frequency counts and contingency tables.
 3. df.set_index() -> Assigns one or more columns as index(multiIndex possible).
     * Needed before hierarchical operations.
 4. df.stack() -> Pivots columns into rows(longer format).
     * Good for reshaping wide->long
 5. df.unstack() -> Pivots rows into columns(wider format).
     * Good for reshaping long->wide
 6. pd.melt() -> Unpivots a DF from wide to long format.
     * Useful for tidy data principles.
 7. df.pivot() -> Reshapes data based on column values.
     * Simpler than pivot_table,no aggregation.
 8. df.explode() -> Transforms list-like entries into separate rows.
     * Useful for JSON or nested data.


Elements & Parameters
 1. values -> Column(s) to aggregate in pivot/crosstab.
 2. margins -> Adds totals row/column.
 3. normalize -> Normalize counts in crosstab.
 4. level -> MultiIndex level to stack/unstack.
