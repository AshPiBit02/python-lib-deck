1. rolling()
   * Apply moving window calculations over fixed rows or time spans.
   * Use cas: Moving averages, volatility, smoothing trends.
   * Essential elements: 
      * window(int or offset like "7D")
      * min_periods(minimum observations required)
      * center=True(centered window)
      * win_type(Gaussian, triangular,etc).
      * Methods: .mean(),.sum(),.std(),.apply(func).

2. expanding()
    * Growing window from the start of the series.
    * Use case: Cumulative statistics that update as more data arrives.
    * Essential elements: 
       * min_periods(start after certain rows)
       * Methods: .sum(),.mean(),.apply(func)


3. Cumulative Functions
    * Track totals or averages across the entire sequence.
    * Use case: Runnign totals, cumulative averages, progress tracking.
    * Essential methods:
        * .cumsum() -> cumulative sum
        * .cumprod() -> cumulative product
        * .cummax()/.cummin() -> cumulative extrema
        * .cummean()(via expanding + mean)

4. Weighted Windows
    * Apply statistical weights to rolling windows.
    * Use case: Smooth data with emphasis on recent or central values.
    * Essential elements:
        * win_type="gaussian"(or "triang","boxer",etc.)
        * Extra parameters like std for Gaussian
        * Methods: .mean(),.apply(func)

5. GroupBy + Window
    * Apply rolling/expanding inside groups.
    * Use case: Regional moving totals.
    * Elemential methods:
        * grouby().tansform(lambda x: x.rolling(...).mean())
        * groupby().cumsum()

6. Lag Analysis(shift)
    * Compare current values with past/future values.
    * Use case: Forecasting, anomaly detection, growth rate calculations.
    * Essential methods:
        * .shift(n) -> move values by n periods.
        * Combine with rolling: df["lagged_avg"]=df["Sales"].shift(1).rolling(3).mean()

7. Custom Functions with .apply()
    * Apply arbitary logic to rolling/expanding windows.
    * Use case: Z-scores, custom anomly detection, domain-specific metrics.
    * Essential methods: 
        * .rolling(....).apply(func,raq=False)
        * .expanding().apply(func)

# Demo Implementation

```python
import pandas as pd
import numpy as np

# Generate 500 rows of daily sales data for 4 regions
dates = pd.date_range("2024-01-01", periods=500, freq="D")
regions = np.random.choice(["North","South","East","West"], size=500)
sales = np.random.randint(100, 1000, size=500)
discounts = np.random.randint(5, 30, size=500)

df = pd.DataFrame({
    "Date": dates,
    "Region": regions,
    "Sales": sales,
    "Discount": discounts
})
df=df.set_index("Date")

print(df.head())
```

```python
# Compute a 7-day moving average of sales for each region.
df["7d_avg_sales"]=df.groupby("Region")["Sales"].transform(lambda x: x.rolling("7D",on=df["Date"]).mean())
print(df.head())
```
