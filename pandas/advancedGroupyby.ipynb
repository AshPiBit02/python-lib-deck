1. groupby().transform()
   * Apply a function to each group but return results with the same shape as the orginal DF.
   * Use case: Normalizing values within groups, computing z-scores, or scaling data per group.

```python
# Example
df["normalized"]=df.groupby("Region")["Sales"].transform(lambda x: x/x.mean())
```

2. groupby().filter()
   * Keep or drop entire groups based on a condition.
   * Use case: Select groups that meet a threshold 

```python
# Example
filtered=df.groupby("Region").filter(lambda x:x["sales"].sum()>1000)
```

3. groupby().apply()
   * Apply a costom function to each group and return combined results.
   * Use case: Complex operations like returning top N rows per group, reshaping, or custom calculations.

```python
# Example
top_sales=df.groupby("Region").apply(lambda x: x.nlargest(2,"sales"))
```

4. groupby().agg()
   * Perform multiple aggregations at once, including custom ones.
   * Use case: Summarize groups with different metrics(sum,mean,max,custom range)

```python
# Example
summary=df.groupby("Region")["Sales"].agg(total="sum",avg="mean",max_sale="max",range=lambda x : x.max()-x.min())
```

5. groupby().size() vs groupby().count()
   * Count rows per group
   * Difference: 
      * .size() -> counts all rows(including NaN)
      * .count() -> counts only non-NaN values.

```python
# Example
df.groupby("Region").size()
df.groupby("Region")["Sales"].count()
```

6. groupby(....,sort=False)
   * Prevent automatic sorting of group keys.
   * Use case: Keep original order of groups.

```python
# Example
df.groupby("Region",sort=False)["Sales"].sum()
```

7. groupby(....,observed=True)
   * Optimize grouping when using categorical columns.
   * Use case: Reduce memory usage and speed up groupby with categorical dype.

```python
# Example
df.groupby("Region",observed=True)["Sales"].mean()
```
