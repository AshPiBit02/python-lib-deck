```python
import pandas as pd

data = {
    "date": pd.date_range("2026-04-01", periods=10, freq="D"),
    "sales": [200, 220, 250, 270, 300, 280, 260, 310, 330, 350]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Convert the date column to datetime using pd.to_datetime()
df["date"]=pd.to_datetime(df["date"])
print(df)
```

```python
# Extract year from the date column.
df["year"]=df["date"].dt.year
print(df)
```

```python
# Extract month from the date column.
df["month"]=df["date"].dt.month
print(df)
```

```python
df["day_of_week"]=df["date"].dt.day_name()
print(df)
```

# Resampling (Groups data into fixed buckets(weeks,months,etc))

```python
import pandas as pd

# Create 30 days of sales data
data = {
    "date": pd.date_range("2026-04-01", periods=30, freq="D"),
    "sales": [200, 220, 250, 270, 300, 280, 260, 310, 330, 350,
              360, 370, 390, 400, 420, 410, 430, 440, 460, 470,
              480, 490, 500, 520, 530, 540, 550, 560, 570, 580]
}

df = pd.DataFrame(data)
df = df.set_index("date")   # set date as index for resampling
print(df.head())
```

```python
# Weekly totals
df.resample("W").sum()
```

```python
# Monthly averages
df.resample("ME").mean()
```

```python
# Resample weekly but change the week ending day to Friday.
df.resample("W-FRI").sum()
```

# Monthly Aggregation

```python
# Resample monthly and calulate th average sales.
df.resample("ME").mean()
```

```python
# Resample monthly and find the maximum sales values.
df.resample("ME").max()
```

```python
# Resample monthly and count the number of records.
df.resample("ME").count()
```

# Custom Intervals

```python
# Resample every 3 days and calculate the sum.
df.resample("3D").sum()
```

```python
# Resample every 10 days and find the maximun.
df.resample("10D").max()
```

# Comparing Aggregations

```python
# Compare weekly sum vs mean side by side
weekly_sum=df.resample("W").sum()
weekly_mean=df.resample("W").mean()
sample=pd.concat([weekly_sum,weekly_mean],axis=1)
sample.columns=["Weekly Sum","Weekly Mean"]
print(sample)
```

```python
# Compare monthly max vs min.
monthly_max=df.resample("ME").max()
monthly_min=df.resample("ME").min()
result=pd.concat([monthly_max,monthly_min],axis=1)
result.columns=["Monthly Max","Monthly Min"]
print(result)
```

# Advanced Twist

```python
# Resample weekly and caculate the standard deviation of sales.
weekly_std=df.resample("W").std()
print(weekly_std)
```

```python
# Resample every 3 days and calculate the cumulative sum.
three_day_cumsum=df.resample("3D").sum().cumsum()
print(three_day_cumsum )
```

```python
# Resample monthly and calculate the median sales.
print(df.resample("ME").median())
```

# Rolling (creates a moving window over data and applies functions inside that window)
  * General Syntax: df["sales"].rolling(window,min_periods=1).aggregation()

```python
import pandas as pd
import numpy as np

# Generate 75 days (~2.5 months) of daily data
dates = pd.date_range("2026-03-01", periods=75, freq="D")

# Create some synthetic temperature values with variation
np.random.seed(42)  # reproducible
temps = np.random.randint(14, 30, size=75)  # temperatures between 14–29 °C

df = pd.DataFrame({"date": dates, "temperature": temps})
df = df.set_index("date")

print(df.head(10))
```

```python
# Compute a 7-day moving average of temperature.
result=df["temperature"].rolling(7).mean().to_frame("7D mean Temp")
print(result[5:18])
```

```python
# Compare the 7-Day vs 14-Day moving averages side by side.
seven_day_avg=df["temperature"].rolling(7).mean()
fourteen_day_avg=df["temperature"].rolling(14).mean()
result=pd.concat([seven_day_avg,fourteen_day_avg],axis=1)
result.columns=["7 Day Mean Temp","14 Day Mean Temp"]
print(result.head(15))
```

```python
# Compare rolling sum with cumulative sum
sum=df["temperature"].rolling(10).sum()
cumsum=df["temperature"].rolling(10).sum().cumsum()
result=pd.concat([sum,cumsum],axis=1)
result.columns=["Sum","Cumsum"]
print(result[10:30])
```

```python
# Find the maximum temperature in the last 5 days.
print(df["temperature"].rolling(5).max().to_frame("5D Rolling Max"))
```

```python
# Find the minimun temperature in the last 10 days.
print(df["temperature"].rolling(10).min().to_frame("10D Rolling Min"))
```

```python
# Compare rolling max vs rolling min
min=df["temperature"].rolling(2).min().to_frame("2D Rolling Min")
max=df["temperature"].rolling(2).max().to_frame("2D Rolling Max")
result=pd.concat([min,max],axis=1)
print(result.tail(12))
```
