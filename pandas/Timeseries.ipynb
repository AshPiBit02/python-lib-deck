```python
import pandas as pd

data = {
    "date": pd.date_range("2026-04-01", periods=10, freq="D"),
    "sales": [200, 220, 250, 270, 300, 280, 260, 310, 330, 350]
}

df = pd.DataFrame(data)
print(df)
```

# Datetime Conversion

```python
# Convert the date column to datetime using pd.to_datetime()
df["date"]=pd.to_datetime(df["date"])
print(df)
```

```python
# Extract year from the date column.
df["year"]=df["date"].dt.year
print(df)
```

```python
# Extract month from the date column.
df["month"]=df["date"].dt.month
print(df)
```

```python
df["day_of_week"]=df["date"].dt.day_name()
print(df)
```

# Date Indexing

```python
# Set date as the index of the DataFrame.
df=df.set_index("date")
print(df)
```

```python
# Select all rows between "2026-04-03" and "2026-04-07"
df.loc["2026-04-03":"2026-04-07"]
```

```python
# Select sales data for "2026-04-05" only.
df.loc["2026-04-05"]
```
