# Components

1. Data 
   * The actual values(numbers, strings, etc).
2. Index
   * Row labels(default: 0,1,2,3,.. but can be custom).
3. Columns
   * Column labels(names of varibles).
4. dtypes
   * Data types of each column.
5. Shape
   * Dimensions(rows x columns).
6. Axes
   * Both row and column labels.

# Essential Functions & Methods

1. Creation & Inspecton
  * pd.DataFrame(data) -> create DataFrame.
  * .head(n)/.tail(n) -> view first/last rows.
  * .info() -> summary of structure
  * .shape() -> (rows,columns)
  * .columns/.index -> labels
  * .dtypes -> column data types.
2. Descriptive Statistics
  * .describe() -> apply custom function
  * .map(func) -> element-wise mapping.
  * .assign() -> add new columns
  * .rename() -> rename columns/index.
3. Selection & Indexing
  * df["col"]/df[["col1","col2"]] -> select columns.
  * df.loc[row_label,col-label] -> label-based selection
  * df.iloc[row_index,col_index] -> position-based selection
  * df[df["col"]>100] -> Boolean indexing
4. Transformation
  * .apply(func) -> apply custom function
  * .map(func) -> element-wise mapping
  * .assign() -> add new columns
  * .rename() -> rename columns/index
5. Handling Missing Data
  * .isnull() / .notnull()
  * .dropna() -> remove missing
  * .fillna(value) -> replace missing
6. Sorting & Ranking
  * .sort_values(by="col")
  * .sort_index()
  * .rank()
7. Combining & Reshaping
  * .concat([df1,df2])
  * .merge(df1,df2,on="col")
  * .join(df2)
  * .pivot_table()
  * .groupby("col")
8. Input/Output
  * pd.read_csv("file.csv")
  * df.to_csv("file.csv")
  * pd.read_excel(),df.to_excel()
  * pd.read_sql()
