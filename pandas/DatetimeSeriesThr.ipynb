# General Syntax: pd.date_range(start=None,end=None,periods=None,freq='D')
* Parameters:
  1. Start -> starting date(string or datetime)
  2. end -> ending date(string or datetime)
  3. periods -> number of dates to generate(alternative to end)
  4. freq -> frequency of the dates(default 'D'=daily)
* Frequency accepts many codes:
  1. 'D' -> day
  2. 'W' -> week
  3. 'M' -> month end
  4. 'MS' -> month start
  5. 'Q' -> quarter end
  6. 'A' -> year end
  7. 'H' -> hour
  8. 'T' or 'min' -> minute
  9. 'S' -> second

# Core Datetime Functions
  1. pd.to_datetime() -> converts strings, lists, or columns into datetime objects
