# General Syntax: pd.date_range(start=None,end=None,periods=None,freq='D')
* Parameters:
  1. Start -> starting date(string or datetime)
  2. end -> ending date(string or datetime)
  3. periods -> number of dates to generate(alternative to end)
  4. freq -> frequency of the dates(default 'D'=daily)
* Frequency accepts many codes:
  1. 'D' -> day
  2. 'W' -> week
  3. 'M' -> month end
  4. 'MS' -> month start
  5. 'Q' -> quarter end
  6. 'A' -> year end
  7. 'H' -> hour
  8. 'T' or 'min' -> minute
  9. 'S' -> second

# Core Datetime Functions
  1. pd.to_datetime() -> converts strings, lists, or columns into datetime objects.
  2. pd.date_range() -> generates a fixed-frequency sequence of dates.
  3. pd.Timestamp() -> Represents a single point in time.
  4. pd.Timedelta() -> Represents a duration(difference between two dates/times).
  5. pd.Period() -> Represents a span of time(month,quarter,year).

# .dt Accessor Attributes
  1. .dt.year -> extract year
  2. .dt.month -> extract month number
  3. .dt.day -> extract day of month
  4. .dt.hour,.dt.minute,.dt.second -> extract time components.
  5. .dt.day_name() -> Day of week name
  6. .dt.month_name() -> Month name
  7. .dt.dayofweek() -> Day of week number(Mon=0).
  8. .dt.dayofyear() -> Day of Year
  9. .dt.is_leap_year() -> Boolean check for leap year
  10. .dt.days_in_month -> Numbers of days in month

# Common Datetime Methods
  1. Formatting & Parsing
    * .strftime("%Y-%m-%d") -> format datetime as string.
    * .normalize() -> set time to midnight
  2. Arithmetic
    * timestamp + pd.Timedelta("1day") -> add one day
    * timestamp + pd.offsets.BDay() -> add one business day
  3. Resampling & Rolling
    * .resample("W").sum() -> weekly aggregation
    * .rolling(3).mean() -> 3-period rolling average
  4. Time Zone Handling
    * .tz_localize("UTC") -> assign timezone
    * .tz_convert("Asia/Kathmandu") -> convert timezone.
