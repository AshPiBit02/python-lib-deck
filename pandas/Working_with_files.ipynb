# Writing Files

# Integration 

```python
# Merge the Sheet1 and Sheet2 sheets of sales.xlsx into a single DataFrame using "id" as the key.
q1=pd.read_excel("dataFiles/sales.xlsx",sheet_name="Sheet1")
q2=pd.read_excel("dataFiles/sales.xlsx",sheet_name="Sheet2")
# merged=pd.merge(q1,q2,on="id")
merged=pd.concat([q1,q2],ignore_index=True)
print(merged)
```

```python
# Save the merged DataFrame into JSON format with orient="records" and indentaion.
merged.to_json("dataFiles/combined_sales.json",orient="records",indent=2)
```
