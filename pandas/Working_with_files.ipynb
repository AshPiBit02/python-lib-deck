# Reading Files

```python
# Load students.csv into a DataFrame. Show the first 3 rows.
import pandas as pd
df=pd.read_csv("dataFiles/students.csv")
print(df.head(3))
```

```python
# Read students.csv but only include the columns name and city.
df=pd.read_csv("dataFiles/students.csv",usecols=["name","city"])
print(df)
```

```python
# Skip the first row of students.csv while reading.
df=pd.read_csv("dataFiles/students.csv",skiprows=1)
print(df)
```

```python
# Load students.csv and set "id" as the index column.
df=pd.read_csv("dataFiles/students.csv")
df=df.set_index("id")
print(df)
```

```python
# Read sales.xlsx and load the sheet 'Sheet1'.
# xls=pd.ExcelFile("dataFiles/sales.xlsx")
# print(xls.sheet_names)
df=pd.read_excel("dataFiles/sales.xlsx",sheet_name="Sheet1")
print(df.head())
```

```python
# Read sailes.xlsx and load the sheed "Sheet2"
df=pd.read_excel("dataFiles/sales.xlsx",sheet_name="Sheet2")
print(df)
```

```python
# Load a JSON file students.json (assume same structure as CSV)
df=pd.read_json("dataFiles/students.json")
print(df)
```

# Writing Files

```python
# Save the students.csv DataFrame into a new file output.csv without the index.
df=pd.read_csv("dataFiles/students.csv")
df.to_csv("dataFiles/output.csv",index=False)
```
