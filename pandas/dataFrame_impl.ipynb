```python
# DataFram from a dictionary
import pandas as pd
data={
    "Name":["Ram","Sita","Hari","Gita","Bina"],
    "Age":[20,21,19,22,20],
    "Marks":[85,90,76,88,95],
    "Department":["CS","IT","CS","ECE","IT"]
}
df=pd.DataFrame(data)
print(df)
```

```python
# DataFrame from a list of lists with column names.
import pandas as pd
data=[
    ["Ram",20,85,"CS"],
    ["Sita",21,90,"IT"],
    ["Hari",19,76,"CS"],
    ["Gita",22,88,"ECE"],
    ["Bina",20,95,"IT"]
]
df=pd.DataFrame(data,columns=["Name","Age","Marks","Department"])
print(df)
```

```python
import pandas as p
data={
    "Product":["Laptop","Phone","Tablet","Camera","Headphones"],
    "Price":[1200,800,300,600,150],
    "Units_Sold":[50,200,124,88,119,]
}
df=p.DataFrame(data)
```

```python
print(df["Price"].map(lambda x: x*133))
```

```python
print(df["Product"].map(lambda x: x.upper()))
```

```python
df=df.assign(Revenue=df["Price"]*df["Units_Sold"])
print(df)
```

```python
df=df.assign(Disounted_Price=df["Price"]*0.9)
print(df)
```

```python
df=df.rename(columns={"Units_Sold": "Quantity"})
print(df)
```

```python
df=df.rename(columns={"Price":"Unit_Price","Product":"Item"})
print(df)
```

```python
df=df.rename(index={0:"i",1:"ii",2:"iii",3:"iv",4:"v"})
print(df)
```

```python
import pandas as pd

data = {
    "Product": ["Laptop", "Phone", "Tablet", "Camera", "Headphones"],
    "Price": [80000, 12000, 6000, 8000, 1500],
    "Units_Sold": [60, 60, 44, 65, 119],
    "Category": ["Electronics", "Electronics", "Electronics", "Electronics", "Accessories"]
}

df = pd.DataFrame(data)
print(df)
```

```python
product_only=df["Product"]
print(product_only)
```

```python
product_price=df[["Product","Price"]]
print(product_price)
```

```python
tablet_price=df.loc[df["Product"]=="Tablet","Price"]
print(tablet_price)
```

```python
row_column_val=df.iloc[2,1]
print(row_column_val)
```

```python
print(df[df["Price"]>10000])
```

```python
print(df["Price"].apply(lambda x: x*1.1))
```

```python
print(df.assign(Revenue=df["Price"]*df["Units_Sold"]))
```

```python
print(df.rename(columns={"Units_Sold":"Quantity"}))
```

```python
df.loc[2,"Price"]=None #missing value introduced
print(df)
```

```python
print(df.dropna()) # removes rows with missing values
```

```python
df.loc[1,"Category"]=None
print(df.fillna(0)) # replace missing values with 0
```

```python
df2=pd.DataFrame({
    "Product":["Laptop","Phone","Tablet"],
    "Discount":[0.1,0.05,0.2]
})
```

```python
print(pd.concat([df,df2],ignore_index=True)) # concat vertically(continue indices)
print(pd.concat([df,df2],ignore_index=False)) # concat vertically (start indices from 0)
```

```python
print(pd.concat([df,df2],axis=1)) # concat horizontally if 0 then verticall concat (row wise)
```
